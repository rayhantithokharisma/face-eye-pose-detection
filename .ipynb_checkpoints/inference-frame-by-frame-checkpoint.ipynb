# Inference via deployed FastAPI

```python
import cv2
import requests
```

```python
# URL of the FastAPI server
url = "http://localhost:8080/pose-estimate"
url2 = "http://localhost:8080/eye-tracking"

cap = cv2.VideoCapture(0)
font = cv2.FONT_HERSHEY_COMPLEX

blue = (0, 0, 255)
red = (255, 0, 0)
green = (0,128,0)

frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = cap.get(cv2.CAP_PROP_FPS)

video_format = cv2.VideoWriter_fourcc('M','J','P','G')

video_file = 'video_out_gaz_5.avi'
video_out = cv2.VideoWriter(video_file,video_format, fps, (frame_width, frame_height))

while True:
    ret, frame = cap.read()

    if not ret:
        break

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    avg_brightness = gray.mean()
    if avg_brightness > 90:
        _, img_encoded = cv2.imencode('.jpg', frame)
    
        response = requests.post(url, data=img_encoded.tobytes())
        response_gazing = response.json()['is_gazing']
        is_no_face = response.json()['is_no_face']
        is_centered = response.json()['is_centered']
    
        response = requests.post(url2, data=img_encoded.tobytes())
        response = response.json()
        response_gazing_2 = response['is_gazing']
        is_no_face_2 = response['is_no_face']
    
        response_gazing_3 = response_gazing or response_gazing_2
        is_no_face_3 = is_no_face or is_no_face_2
    
        if 'left_pupil' in response:
            left_pupil = response['left_pupil']
            right_pupil = response['right_pupil']
            cv2.circle(frame, (left_pupil[0], left_pupil[1]), 3, blue, 3)
            cv2.circle(frame, (right_pupil[0], right_pupil[1]), 3, red, 3)
    
        cv2.putText(frame, f"is gazing : {response_gazing_3}", (10, 100), font, 0.4, red, 1)
        cv2.putText(frame, f"is no face : {is_no_face_3}", (10, 120), font, 0.4, red, 1)
        cv2.putText(frame, f"is centered : {is_centered}", (10, 140), font, 0.4, red, 1)
    
        if response_gazing:
            cv2.putText(frame, f"Please keep your head pointing to the screen!", (10, 50), font, 0.5, blue, 2)
        if response_gazing_3:
            cv2.putText(frame, f"Please do not looking around!", (10, 20), font, 0.5, blue, 2)
        if not is_centered:
            cv2.putText(frame, f"Please keep your face centered", (10, 180), font, 0.5, blue, 2)
    
        if is_no_face_3:
            cv2.putText(frame, f"Please stay in the screen", (10, 220), font, 0.5, blue, 2)
    else:
        cv2.putText(frame, f"You're camera too dark", (10, 240), font, 0.5, blue, 2)
    

    # frame = cv2.resize(frame, (1280, 720))
    video_out.write(frame) 

    cv2.imshow('Video', frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
video_out.write(frame) 
cv2.destroyAllWindows()
```
